```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('zomato.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
dff = df.drop(['url', 'address', 'phone', 'dish_liked', 'menu_item', 'name', 'location', 'rest_type'], axis = 1)
dff
```

```python
dff.isnull().sum()
```

```python
dff.rate = dff['rate'].fillna(0)
```

```python
dff['rest_type'] = dff['listed_in(type)']
dff['location'] = dff['listed_in(city)']
dff
```

```python
df2 = dff.drop(['listed_in(type)', 'listed_in(city)', 'reviews_list'], axis = 1)
```

```python
df2.isnull().sum()
```

```python
ndf = df2.dropna()
ndf.head()
```

```python
ndf.isnull().sum()
```

```python
ndf.to_csv('Zomato_cleaned.csv')
```

```python
type(ndf.rate)
```

```python
dt = pd.read_excel('Zomato_cleaned.xlsx')
dt.head()
```

```python
dt['online_order'] = dt.online_order.map(dict(Yes = 1, No = 0))
dt
```

```python
dt2 = dt.drop(['online_order', 'cuisines', 'cost_for_two_people', 'rest_type', 'location'], axis = 1)
dt2
```

```python
dt3 = ndf.online_order.map(dict(Yes = 1, No = 0))
dt3
```

```python
dt6 = pd.concat([dt3, dt2], axis = 1, join = 'inner')
dt6
```

```python
dd = dt6.drop(dt6[dt6.rate == 'NEW'].index)
dd
```

```python
dd2 = dd.drop(dd[dd.rate == '-'].index)
dd2
```

```python
dd2.rate = dd2.rate.astype(float)
dd2
```

```python
m = dd2.corr().round(2)
m
```

```python
sns.heatmap(m, annot = True)
plt.show
```
